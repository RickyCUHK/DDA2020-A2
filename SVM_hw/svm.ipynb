```python
import re
import numpy as np
from sklearn import datasets
from sklearn.multiclass import OutputCodeClassifier
from sklearn.svm import LinearSVC
import matplotlib.pyplot as plt
from sklearn import svm, datasets

iris = datasets.load_iris()
setosa_train = (iris["target"] == 0).astype( np.int64 )[10:130]
versicolor_train = (iris["target"] == 1).astype( np.int64 )[10:130]
versicolor_train[80:90] = 0
virginica_train = (iris["target"] == 2).astype( np.int64 )[20:140]
setosa_test = (iris["target"] == 2).astype( np.int64)[0:30]
versicolor_test = (iris["target"] == 2).astype( np.int64)[0:30]
virginica_test = (iris["target"] == 2).astype( np.int64)[0:30]
d_train = (iris["target"] == 2).astype( np.float64)[0:120]
d_test = (iris["target"] == 2).astype( np.float64)[0:30]
setosa_test[0:10] = 1
versicolor_test[10:20] = 1
virginica_test[20:30] = 1
d_train[0:120] = -1e5
d_test[0:30] = -1e5
## buld one-to-many datasets


class SVM(object):

    def __init__(self, training_dataset_, test_dataset_):
        self.training_dataset = training_dataset_
        self.test_dataset = test_dataset_
        self.classes = {}
        self.X_train = None
        self.Y_train = None
        self.X_test = None
        self.Y_test = None

        self.distance_train = None
        self.label_train = None
        self.distance_test = None
        self.label_test = None
        
    def read_data(self):
        f = open(self.training_dataset, 'r')
        rows = list(re.split(' ', row) for row in re.split('\n', f.read())[:-1])
        name, self.Y_train = np.unique(list(row[-1] for row in rows), return_inverse=True)
        name, self.label_train = np.unique(list(row[-1] for row in rows), return_inverse=True)
        self.distance_train = d_train
        self.distance_train[0:120] = -1e5
        self.X_train = np.empty((0,4), float)
        f.close()
        for row in rows:
            self.X_train = np.append(self.X_train, np.array([np.array(row[:-1]).astype(float)]), axis = 0)
        f = open(self.test_dataset, 'r')
        f.close()
        
        f = open(self.test_dataset, 'r')
        rows = list(re.split(' ', row) for row in re.split('\n', f.read())[:-1])
        name, self.Y_test = np.unique(list(row[-1] for row in rows), return_inverse=True)
        name, self.label_test = np.unique(list(row[-1] for row in rows), return_inverse=True)
        self.distance_test = d_test
        self.distance_test[0:30] = -1e5
        self.X_test = np.empty((0,4), float)
        f.close()
        for row in rows:
            self.X_test = np.append(self.X_test, np.array([np.array(row[:-1]).astype(float)]), axis = 0)

    def score(self, clf, i):
        count = 0
        for j in self.X_train:
            distance = clf.decision_function(j.reshape(1,-1))
            if (distance[0]) > self.distance_train[count]:
                self.distance_train[count] = distance
                self.label_train[count] = i
            count = count+1
        count = 0
        for k in self.X_test:
            distance = clf.decision_function(k.reshape(1,-1))
            if (distance[0]) > self.distance_test[count]:
                self.distance_test[count] = distance
                self.label_test[count] = i
            count = count+1

    def calculate_error(self):
        train_count = 0
        test_count = 0
        for i in range(120):
            if self.Y_train[i] != self.label_train[i]:
                train_count += 1
        for i in range(30):
            if self.Y_test[i] != self.label_test[i]:
                test_count += 1
        train_error = train_count/120
        test_error = test_count/30
        return train_error, test_error
            
    def SVM(self, i):
    
        #########################
        ## WRITE YOUR CODE HERE##
        #########################    
        clf = svm.SVC(C = 1e5, kernel='linear')

        #train setosa
        if i == 0: 
            clf.fit(self.X_train, setosa_train)
            self.score(clf, 0)
            support_vectors = clf.support_
            w = clf.coef_
            b = clf.intercept_
        elif i == 1:
            clf.fit(self.X_train, versicolor_train)
            self.score(clf, 1)
            support_vectors = clf.support_
            w = clf.coef_
            b = clf.intercept_
        elif i == 2:
            clf.fit(self.X_train, virginica_train)
            self.score(clf, 2)
            support_vectors = clf.support_
            w = clf.coef_
            b = clf.intercept_
        else:
            print("invalid input")

        return support_vectors, w, b
    
    def SVM_slack(self, c, i):
    
        #########################
        ## WRITE YOUR CODE HERE##
        #########################    
        
        clf = svm.SVC( C = c,kernel='linear')

        #train setosa
        if i == 0: 
            clf.fit(self.X_train, setosa_train)
            self.score(clf, 0)
            support_vectors = clf.support_
            w = clf.coef_
            b = clf.intercept_
        elif i == 1:
            clf.fit(self.X_train, versicolor_train)
            self.score(clf, 1)
            support_vectors = clf.support_
            w = clf.coef_
            b = clf.intercept_
        elif i == 2:
            clf.fit(self.X_train, virginica_train)
            self.score(clf, 2)
            support_vectors = clf.support_
            w = clf.coef_
            b = clf.intercept_
        else:
            print("invalid input")

        return support_vectors, w, b
    
    def SVM_kernel_poly2(self, c, i):
    
        #########################
        ## WRITE YOUR CODE HERE##
        #########################    
        clf = svm.SVC(C = c, kernel='poly', gamma=1, degree=2)

        #train setosa
        if i == 0: 
            clf.fit(self.X_train, setosa_train)
            self.score(clf, 0)
            support_vectors = clf.support_
            b = clf.intercept_
        elif i == 1:
            clf.fit(self.X_train, versicolor_train)
            self.score(clf, 1)
            support_vectors = clf.support_
            b = clf.intercept_
        elif i == 2:
            clf.fit(self.X_train, virginica_train)
            self.score(clf, 2)
            support_vectors = clf.support_
            b = clf.intercept_
        else:
            print("invalid input")

        return support_vectors, b
    
    def SVM_kernel_poly3(self, c, i):
    
        #########################
        ## WRITE YOUR CODE HERE##
        #########################    
        
        clf = svm.SVC(C = c, kernel='poly', gamma=1, degree=3)

        #train setosa
        if i == 0: 
            clf.fit(self.X_train, setosa_train)
            self.score(clf, 0)
            support_vectors = clf.support_
            b = clf.intercept_
        elif i == 1:
            clf.fit(self.X_train, versicolor_train)
            self.score(clf, 1)
            support_vectors = clf.support_
            b = clf.intercept_
        elif i == 2:
            clf.fit(self.X_train, virginica_train)
            self.score(clf, 2)
            support_vectors = clf.support_
            b = clf.intercept_
        else:
            print("invalid input")

        return support_vectors, b

    def SVM_kernel_rbf(self, c, i):
    
        #########################
        ## WRITE YOUR CODE HERE##
        #########################    
        clf = svm.SVC(C = c, kernel='rbf', gamma=0.5)

        #train setosa
        if i == 0: 
            clf.fit(self.X_train, setosa_train)
            self.score(clf, 0)
            support_vectors = clf.support_
            b = clf.intercept_
        elif i == 1:
            clf.fit(self.X_train, versicolor_train)
            self.score(clf, 1)
            support_vectors = clf.support_
            b = clf.intercept_
        elif i == 2:
            clf.fit(self.X_train, virginica_train)
            self.score(clf, 2)
            support_vectors = clf.support_
            b = clf.intercept_
        else:
            print("invalid input")

        return support_vectors, b

    
    def SVM_kernel_sigmoid(self, c, i):
    
        #########################
        ## WRITE YOUR CODE HERE##
        #########################    
        
        clf = svm.SVC(C = c, kernel='sigmoid', gamma='auto', coef0=0)

        #train setosa
        if i == 0: 
            clf.fit(self.X_train, setosa_train)
            self.score(clf, 0)
            support_vectors = clf.support_
            b = clf.intercept_
        elif i == 1:
            clf.fit(self.X_train, versicolor_train)
            self.score(clf, 1)
            support_vectors = clf.support_
            b = clf.intercept_
        elif i == 2:
            clf.fit(self.X_train, virginica_train)
            self.score(clf, 2)
            support_vectors = clf.support_
            b = clf.intercept_
        else:
            print("invalid input")

        return support_vectors, b
```

```python
train = SVM("train.txt", "test.txt")

def outPut(file, training_error, testing_error ,support_vector0, w0, b0,\
        support_vector1, w1, b1, \
        support_vector2, w2, b2):

    fo = open(str(file), "w")
    fo.write(str(training_error)+"\n")
    fo.write(str(testing_error)+"\n")
    for x in w0:
        fo.write(",".join(str(y) for y in x)+"\n")
    fo.write(",".join(str(y) for y in b0)+"\n")
    fo.write(",".join(str(y) for y in support_vector0)+"\n")
    for x in w1:
        fo.write(",".join(str(y) for y in x)+"\n")
    fo.write(",".join(str(y) for y in b1)+"\n")
    fo.write(",".join(str(y) for y in support_vector1)+"\n")
    for x in w2:
        fo.write(",".join(str(y) for y in x)+"\n")
    fo.write(",".join(str(y) for y in b2)+"\n")
    fo.write(",".join(str(y) for y in support_vector2))
    fo.close()

def outPut_kernel(file, training_error, testing_error,support_vector0, b0,\
        support_vector1, b1, \
        support_vector2, b2):

    fo = open(str(file), "w")
    fo.write(str(training_error)+"\n")
    fo.write(str(testing_error)+"\n")

    fo.write(",".join(str(y) for y in b0)+"\n")
    fo.write(",".join(str(y) for y in support_vector0)+"\n")

    fo.write(",".join(str(y) for y in b1)+"\n")
    fo.write(",".join(str(y) for y in support_vector1)+"\n")

    fo.write(",".join(str(y) for y in b2)+"\n")
    fo.write(",".join(str(y) for y in support_vector2))
    fo.close()

#####################################
## Call different SVM with value C ##
#####################################

##############################
##        Linear SVM        ##
##############################
train.read_data()
support_vector0, w0, b0 = train.SVM(0)
support_vector1, w1, b1 = train.SVM(1)
support_vector2, w2, b2 = train.SVM(2)
train_error, test_error = train.calculate_error()
outPut("SVM_linear.txt", train_error, test_error, support_vector0, w0, b0,\
        support_vector1, w1, b1,\
        support_vector2, w2, b2)


##############################
## SVM with slack variables ##
##############################
for t in range(1,11):
    c = 0.1*t
    c = round(c,1)
    train.read_data()
    support_vector0, w0, b0 = train.SVM_slack(c, 0)
    support_vector1, w1, b1 = train.SVM_slack(c, 1)
    support_vector2, w2, b2 = train.SVM_slack(c, 2)
    train_error, test_error = train.calculate_error()
    outPut("SVM_slack_%s.txt"%c, train_error, test_error, support_vector0, w0, b0,\
        support_vector1, w1, b1,\
        support_vector2, w2, b2)

###########################################
## kernel functions with slack variables ##
###########################################

#################################
## 2nd-order polynomial kernel ##
#################################
train.read_data()
support_vector0, b0 = train.SVM_kernel_poly2(1, 0)
support_vector1, b1 = train.SVM_kernel_poly2(1, 1)
support_vector2, b2 = train.SVM_kernel_poly2(1, 2)
train_error, test_error = train.calculate_error()
outPut_kernel("SVM_poly2.txt", train_error, test_error,
        support_vector0, b0,\
        support_vector1, b1, \
        support_vector2, b2)

#################################
## 3rd-order polynomial kernel ##
#################################
train.read_data()
support_vector0, b0 = train.SVM_kernel_poly3(1, 0)
support_vector1, b1 = train.SVM_kernel_poly3(1, 1)
support_vector2, b2 = train.SVM_kernel_poly3(1, 2)
train_error, test_error = train.calculate_error()
outPut_kernel("SVM_poly3.txt", train_error, test_error,
        support_vector0, b0,\
        support_vector1, b1, \
        support_vector2, b2)

##################################
## Radial Basis Function kernel ##
##################################
train.read_data()
support_vector0, b0 = train.SVM_kernel_rbf(1, 0)
support_vector1, b1 = train.SVM_kernel_rbf(1, 1)
support_vector2, b2 = train.SVM_kernel_rbf(1, 2)
train_error, test_error = train.calculate_error()
outPut_kernel("SVM_rbf.txt", train_error, test_error,
        support_vector0, b0,\
        support_vector1, b1, \
        support_vector2, b2)

##################################
##       Sigmoidal kernel       ##
##################################
train.read_data()
support_vector0, b0 = train.SVM_kernel_sigmoid(1, 0)
support_vector1, b1 = train.SVM_kernel_sigmoid(1, 1)
support_vector2, b2 = train.SVM_kernel_sigmoid(1, 2)
train_error, test_error = train.calculate_error()
outPut_kernel("SVM_sigmoid.txt", train_error, test_error,
        support_vector0, b0,\
        support_vector1, b1, \
        support_vector2, b2)
```
